```python
df2 = pd.read_excel('alimenticia_senai.xlsx', sheet_name='orcamento')
df2
```

```python
df2['setor'].value_counts()
```

```python
def num_para_mes(num):
    match num:
        case 1:
            return 'Jan'
        case 2:
            return 'Fev'
        case 3:
            return 'Mar'
        case 4:
            return 'Abr'
        case 5:
            return 'Mai'
        case 6:
            return 'Jun'
        case 7:
            return 'Jul'
        case 8:
            return 'Ago'
        case 9:
            return 'Set'
        case 10:
            return 'Out'
        case 11:
            return 'Nov'
        case 12:
            return 'Dez'

df2['mes'] = df2['mes'].map(num_para_mes)
df2
```

```python
df2['ano']
```

```python
df2['valor_previsto']
```

```python
def duas_casas(num):
    formatado = f'{num:.2f}'
    return float(formatado)

df2['valor_realizado'] = df2['valor_realizado'].map(duas_casas)
df2
```

```python
df2.dropna()
```

```python
with pd.ExcelWriter('alimenticia_senai_tratado.xlsx') as writer:
    df.to_excel(writer, sheet_name='despesas', index=False)
    df2.to_excel(writer, sheet_name='orcamento', index=False)
```
