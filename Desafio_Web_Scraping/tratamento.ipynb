```python
import pandas as pd
import numpy as np

df = pd.read_excel('imoveis_vivareal_scraping.xlsx')
df
```

```python
# Quartos

# Remove o texto
df['quartos'] = df['quartos'].str.replace('Quantidade de quartos\n', '').str.replace('2-3', '3').str.strip()

# Transforma para float
df['quartos'] = df['quartos'].apply(float)
df
```

```python
# Metragem

# Remove as partes com texto
df['metragem'] = df['metragem'].str.replace('Tamanho do imóvel\n', '').str.replace('m²', '').str.strip()

# Substitui a linha 45 de '60-63' para '62' (proximo da media)
df.loc[45, 'metragem'] = 62

# Transforma tudo para float
df['metragem'] = df['metragem'].apply(float)
df
```

```python
# Banheiros

# Remove o texto
df['banheiros'] = df['banheiros'].str.replace('Quantidade de banheiros\n', '').str.replace('1-2', '2').str.replace('2-3', '3').str.strip()

# Transforma para float
df['banheiros'] = df['banheiros'].apply(float)
df
```

```python
# Vagas

# Remove o texto
df['vagas'] = df['vagas'].str.replace('Quantidade de vagas de garagem\n', '').str.replace('1-2', '2').str.replace('0-3', '3').str.strip()

# Transforma para float
df['vagas'] = df['vagas'].apply(float)
df
```

```python
# Valor

# Remove o cifrao
df['valor'] = df['valor'].str.replace('R$', '').str.replace('A partir de ', '').str.replace('.', '').str.strip()

# Transforma para float
df['valor'] = df['valor'].apply(float)
df 
```

```python
# IPTU

# Remove o cifrao
df['iptu'] = df['iptu'].str.replace('R$', '').str.strip()

# Transforma para float
df['iptu'] = df['iptu'].apply(float)
df
```

```python
# Condominio
df['condominio'] = df['condominio'].str.replace('R$','').str.replace('.', '').str.replace('a partir de ', '').str.strip()

# Transfomar para float
df['condominio'] = df['condominio'].apply(float)
df
```

```python
# Excel

df.to_excel('imoveis_vivareal_tratado.xlsx', index=False)
```
