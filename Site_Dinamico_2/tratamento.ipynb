```python
import pandas as pd
import numpy as np

df = pd.read_excel('placas_de_video.xlsx')
df
```

```python
# Transformar a coluna de preco em float

def para_float(preco):
    preco = preco.strip().replace('R$ ', '').replace('.', '').replace(',', '.')
    return float(preco)

df['preco'] = df['preco'].apply(para_float)
df
```

```python
# Memória

memorias = {
    '24': ['24 GB', '24GB', '24 G', '24G'],
    '20': ['20 GB', '20GB', '20 G', '20G'],
    '16' : ['16 GB', '16GB', '16 G', '16G'],
    '12': ['12 GB', '12GB', '12 G', '12G'],
    '8': ['8 GB', '8GB', '8 G', '8G'],
    '4': ['4 GB', '4GB', '4 G', '4G'],
    '2': ['2 GB', '2GB', '2 G', '2G', '2048 MB', '2048MB'],
    '1': ['1 GB', '1GB', '1 G', '1G', '1024 MB', '1024MB']
}

def identificar_memoria(titulos):
    titulos = titulos.upper().strip()
    
    for memoria, variantes in memorias.items():
        for variante in variantes:
            if variante in titulos:
                return memoria
    return np.nan
    
df['memoria (GB)'] = df['titulo'].apply(identificar_memoria)
df['memoria (GB)'].value_counts(dropna=False)
```

```python
# GDDR

gddrs = {
    '3': ['DDR3'],
    '4': ['DDR4'],
    '5': ['DDR5', 'DD5'],
    '6': ['DDR6']
}


def identificar_ddr(titulos):
    titulos = titulos.upper().strip()
    
    for ddr, variantes in gddrs.items():
        for variante in variantes:
            if variante in titulos:
                return ddr
    return np.nan
   
df['gddr'] = df['titulo'].apply(identificar_ddr)

df['gddr'] = df['gddr'].str.replace('DDR', '').str.strip().str.replace('DD', '')
```

```python
# Marcas

marcas = {
    'AFOX': ['AFOX'],
    'AITEK': ['AITEK'],
    'ASROCK': ['ASROCK'],
    'ASUS': ['ASUS'],
    'BLUECASE': ['BLUECASE'],
    'BRX': ['BRX'],
    'DEX': ['DEX'],
    'DUEX': ['DUEX'],
    'GIGABYTE': ['GIGABYTE'],
    'OUTROS': ['OUTROS'],
    'PCWINMAX': ['PCWINMAX'],
    'PCYES': ['PCYES', 'PC YES'],
    'POWERCOLOR': ['POWERCOLOR', 'POWER COLOR'],
    'SAPPHIRE': ['SAPPHIRE'],
    'VALIANTY': ['VALIANTY'],
    'VXPRO': ['VXPRO', 'VX PRO'],
    'XFX': ['XFX']
}

def identificar_marcas(titulos):
    titulos = titulos.upper().strip()

    for marca, variantes in marcas.items():
        for variante in variantes:
            if variante in titulos:
                return marca
    return np.nan

df['marca'] = df['titulo'].apply(identificar_marcas)
df
    
```
