```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('placas_de_video_tratada.xlsx')
df
```

```python
# Media de preco por modelo e marca

df.groupby(['marca', 'modelo'])['preco'].mean()
```

```python
# Visualizacao

tabela_pivot = df.pivot_table(index='modelo', columns='marca', values='preco', aggfunc='mean')

plt.figure(figsize=(10,5))
sns.heatmap(tabela_pivot, annot=True, fmt='.0f')
plt.title('Preço médio por modelo e marca')
plt.show()
```

```python
# preço médio por marca

df.groupby('marca')['preco'].mean().sort_values(ascending=False)
```

```python
# Visualizacao

media_marca = df.groupby('marca')['preco'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
media_marca.plot(kind='barh', color='skyblue')
plt.xlabel('Média de Preço (R$)')
plt.ylabel('Marcas')
plt.grid()
plt.show()
```

# Análise

## 1. Variedade de ofertas

A marca XFX tem a maior variedade de modelos intermediários e potentes de placa de video.
E logo atrás, a ASUS e a ASROCK acompanham.

A PCYES tem a maior variedade de modelos antigos de placa de videos.

## 2. Média de Preços por Marca

As marcas ASROCK, ASUS, GIGABYTE, SAPPHIRE e XFX tem mas maiores médias de preço. Isso se deve ao fato de venderem modelos potentes e recentes com uma maior quantidade. 

As marcas com as médias de preço menores vendem modelos antigos, menos potentes e com uma menor ou igual quantidade.

## 3. Conclusão

Para quem busca placas de videos antigas e baratas, as marcas AFOX, PCWINMAX e PCYES têm bons preços e boas variedades.
Agora, para quem busca placas de videos recentes e potentes, portanto mais caras, as marcas XFX, ASUS e ASROCK possuem preços competitivos e uma diversidade de modelos.
